# Q1

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)
df = dfs[0][1:]; df.columns=dfs[0].loc[0,:]
```

### Delete the line containing 'Not assigned' in Neighbourhood
### Convert column names to distributed csv format

```python
df.drop(df.index[df.Neighbourhood == 'Not assigned'], inplace=True)
df = df.rename(columns={'Postcode': 'Postal Code'})
```
